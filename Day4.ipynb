````python
# Example 4: Conditions with logical operators (and, or, not)
has_ticket = bool(input())
has_id = bool(input())
print(f"Has Ticket: {has_ticket}, Has ID: {has_id}")

if has_ticket and has_id:
    print("Entry allowed.")
elif has_ticket or has_id:
    print("Partial requirements met. Check with staff.")
    if not has_id:
        print("Warning: ID is missing!")
   
else:
    print("Entry denied.")
````

## 2. Looping Statements

Loops are used to execute a block of code repeatedly.

### 2.1 `for` Loops

`for` loops are used to iterate over a *sequence* (like a list, tuple, dictionary, set, or string) or other *iterable* object.

**Syntax:**

```python
for variable in iterable:
    # Code block to execute for each item in the iterable
    # 'variable' takes the value of the current item in each iteration
    print(variable)
    

````python
number_list = [1,2,3,4,5,66,77]
strings="""Global web icon

We would like to show you a description"""
````

````python
for a in strings:
    if a in ['a']:
        print("yes",a ,end=',')
    else:
        print(end='')
     
````

````python
for i,y in enumerate(number_list):
    print(f"i: {i}, y: {y}")
````
