```python
a = 15
b = 4

print(f"a = {a}, b = {b}")
print(f"Addition (a + b): {a + b}")           # Addition
print(f"Subtraction (a - b): {a - b}")        # Subtraction
print(f"Multiplication (a * b): {a * b}")     # Multiplication
```

```python
print(f"Division (a/b): {a/b}")
print(f"Floor Division (a//b):{a//b}")
print(f"Modulus (a%b):{a%b}")
```

```python
a=10
b=3
print({a//b})
```

```python
# Modulus operator for remainder
print(f"17 % 5 = {17 % 5}")  

# Exponentiation
print(f"2 ** 3 = {2 ** 3}")  
print(f"5 ** 2 = {5 ** 2}")

# Negative numbers
print(f"-10 // 3 = {-10 // 3}")
print(f"-10 % 3 = {-10 % 3}")
```

```python
# Comparison operators
x = 10
y = 5

print(f"x = {x}, y = {y}")
print(f"Equal to (x == y): {x == y}")
print(f"Not equal to (x != y): {x != y}")
print(f"Greater than (x > y): {x >= y}")
```
