````python
print("Looping from 0 to 4:")
for i in range(5): # 0, 1, 2, 3, 4
    print(i)

print("-" * 20)
````

````python
print("Looping from 1 to 10 with step 2:")
for i in range(1, 11, 2):
    print(i)

print("-" * 20)
````

````python
message = "Hello "
print(len(message))
print("Looping through a string:")
for char in message:
    print(char)

print("-" * 20)
````

````python
fruits = ["apple", "banana", "cherry", "pineapple"]

print("Looping through fruits with index:")
for i,fruit in enumerate(fruits):
    print(f" {i+1}: {fruit}")
#enumerate is the counter to the iterable object.
````

**Syntax:**

```python

    # Code block to execute as long as condition is True
    print("Condition is still True")
    # IMPORTANT: Usually, something inside the loop needs to
